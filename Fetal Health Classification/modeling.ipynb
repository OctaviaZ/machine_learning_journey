```python
import pandas as pd
pd.options.display.max_columns = None
df = pd.read_csv('data/archive.zip')
df
```

```python
target = 'fetal_health'
cols = df.columns.values.tolist()
features = [i for i in cols if i != target]
```

```python
df['at_risk'] = df['fetal_health'].apply(lambda x: 1 if x > 1 else 0)
# logistic regression only works with 2 categories -> define as 0 healthy and 1 at risk
target = 'at_risk'
```

```python
from sklearn.preprocessing import StandardScaler
df[features] = StandardScaler().fit_transform(df[features])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X=df[features], y=df[target])

predictions = model.predict(X=df[features])
predictions
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(df[target], predictions)
```
